```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.Day_1, data.Day_2, test_size = 0.10, random_state = 4)
x_train.shape
```

```python
x_train=x_train.iloc[1:].values
x_train=x_train.reshape(len(x_train),1)
y_train=y_train.iloc[1:].values
y_train=y_train.reshape(len(y_train),1)
x_test=x_test.iloc[1:].values
x_test=x_test.reshape(len(x_test),1)
y_test=y_test.iloc[1:].values
y_test=y_test.reshape(len(y_test),1)
```

```python
def slr_sgd(x,y,epochs):
    length=len(x)
    #print(length)
    i=j=0
    m=1#[1]*(length)
    c=-1#[-1]*(length)
    vm=0
    vc=0
    while(j<epochs):
        while(i<length):
            k=(y[i]-(m*x[i])-c[i])
            dm=-0.01*(k*x[i])
            dc=-0.01*k
            #print(vc)
            vm=(vm*0.9)-dm
            vc=(vc*0.9)-dc
            m=m+vm
            c=+vc
            i=i+1
        j=j+1
    return m,c
    
```

```python
m=1
c=-1
lr=0.01
delta_m=1
delta_c=1
v_m=0
v_c=0
lam=0.9
max_iters=50
iters_count=0
def deriv(m_f,c_f,x,y):
  m_deriv=-1*(y-m_f*x-c_f)*x
  c_deriv=-1*(y-m_f*x-c_f)
  return m_deriv,c_deriv

while iters_count<50:
  for i in range(len(x_train)):
    delta_m,delta_c=deriv(m,c,x_train[i],y_train[i])
    v_m = lam*v_m - lr*delta_m
    v_c=lam*v_c - lr * delta_c
    m=m+v_m
    c=c+v_c
    i=i+1
    iters_count =iters_count +1
    print(f"Iteration: {iters_count}\t value of m: {m},\t value of c:{c}")
```
